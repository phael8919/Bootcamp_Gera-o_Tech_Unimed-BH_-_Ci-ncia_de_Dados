```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Upload do arquivo
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
df.dtypes
```

```python
#Qual a receita total?
df['Valor Venda'].sum()
```

```python
#Qual o custo total?
df['custo'] = df['Custo Unitário'].mul(df['Quantidade']) #Criando a coluna custo
```

```python
df.sample(10)
```

```python
#Qual o custo total?
round(df['custo'].sum(),2)
```

```python
#Qual o lucro total?
round(df['Valor Venda'].sum() - df['custo'].sum(),2)
```

```python
#Criando a coluna lucro
df['Lucro'] = df['Valor Venda'] - df['custo'] 
```

```python
df.sample(5)
```

```python
#Criando colunas com total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.sample(5)
```

# Tempo médio de envio para cada marca

```python
#Extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.sample(5)
```

```python
df['Tempo_envio'].dtypes
```

```python
#Média de tempo de envio por marca
df.groupby('Marca')['Tempo_envio'].mean()
```

# Missing Values

```python
#Verificando se temos valores faltantes
df.isnull().sum()
```

# Lucro por marca

```python
#Vamos agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto")['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Quantidade');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
#Selecionando somente as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head(7)
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro X mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby(['Marca'])['Lucro'].sum().plot.bar(title='Lucro X Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal'); #Exibe o rótulo do eixo x na horizontal
```

```python
df_2009.groupby(['Classe'])['Lucro'].sum().plot.bar(title='Lucro X Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
